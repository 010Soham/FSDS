# 3. What is a box tuple, and how does it work?


<li>Quite a few Pillow functions take a box tuple as argument</li>
<li>This indicates that the respective function is expecting 4 integer coordinates that describe a rectangle</li>
<ul>The order of argumemts is as follows
    <li>Left The x-coordinate of the leftmost edge of the box.</li>
    <li>Top The y-coordinate of the top edge of the box.</li>
    <li>Right The x-coordinate of one pixel to the right of the rightmost edge of the box. This integer must be greater than the left integer.</li>
    <li>Bottom The y-coordinate of one pixel lower than the bottom edge of the box. This integer must be greater than the top integer.</li>
</ul>
<li>box includes the left and top coordinates and goes up to but does not include the right and bottom coordinates.</li>
<li>For example, the box tuple (3, 1, 9, 6) represents all the pixels in the black box</li>

# 5. What method would you call to get Image object for a 100×100 image, excluding the lower-left quarter of it?


```python
from PIL import Image
import numpy as np

#Convert our image to RGBA
img = Image.open(r'sample.png').convert('RGBA')

#Get the imagee data in pur array
array = np.array(img)

#Excluding the lower left quarter by marking the pixels present over there as black
array[446: 893, 0 : 360] = (0, 0, 0,0) 

img = Image.fromarray(array)

img.show()

#saving the new image
img1 = img.save('img_1.png')
```

# 8. Image objects do not have drawing methods. What kind of object does? How do you get this kind of object?


```python
import math 
from PIL import Image, ImageDraw 
  
#Giving dimensions of rectangle
width,height = 200,200
shape = [(100,100),(width - 20, height - 20)]

  
# creating new Image object of the given dimensions 
img = Image.new("RGBA", (width, height)) 
  
# Drawing the rectangle
img1 = ImageDraw.Draw(img)   
img1.rectangle(shape,outline ="red") 
img.show() 
```

<li>PIL.ImageDraw object has Draw()</li>
<li>And this is how we get PIL.ImageDraw objects</li>
